```python
# import pandas as pd
from astropy.time import Time
from astropy.coordinates import SkyCoord, EarthLocation
import astropy.units as u

TTVdfb = pd.read_csv('transit_center_inner.csv')
TTVdfc = pd.read_csv('transit_center_outer.csv')
TTVBbjd = pd.DataFrame(index=range(len(TTVdfb)), columns=['Planet Number', 'Epoch','0.5','0.16','0.84','0.005', '0.995'])
TTVCbjd = pd.DataFrame(index=range(len(TTVdfc)), columns=['Planet Number', 'Epoch','0.5','0.16','0.84','0.005', '0.995'])
```

```python
# Target position
target = SkyCoord(ra="19h05m30.24s", dec="-41d26m15.5s", frame = 'icrs')

# Earth-based observer location (e.g., Keck Observatory)
location = EarthLocation.of_site('Keck Observatory')

# Observation time
for i in range(0, len(TTVdfc)):
    for j in range(1, len(TTVdfc.columns)-1):
        jd = Time(TTVdfc.iloc[i,j], format='jd', scale='utc')
        # Light travel time correction and BJD
        lt_corr = jd.light_travel_time(target, location=location)
        bjd = jd.tdb + lt_corr  # .tdb ensures you're in Barycentric Dynamical Time
        bjdTTV = bjd-2457000
        print("BJD ", TTVdfc.iloc[i,0]," :", bjdTTV)
        TTVCbjd.iat[i, j] = bjdTTV
print(TTVCbjd)
        
```

```python
#for index, rows in TTVdfc.iterrows():
for i in range(0, len(TTVdfb)):
    for j in range(2, len(TTVdfb.columns)-1):
        bjd = TTVdfb.iloc[i, j]
        bjdTTV = bjd-2457000
        #print("BJD ", TTVdfb.iloc[i,0]," :", bjdTTV)
        TTVBbjd.iat[i, j] = bjdTTV
        TTVBbjd.iat[i,1] = TTVdfb.iloc[i,0]
        TTVBbjd.iat[i,0] = 1
print(TTVBbjd)
TTVBbjd.to_csv('transit_center_b_inner_subtract.csv', index=False)
```

```python
#for index, rows in TTVdfc.iterrows():
for i in range(0, len(TTVdfc)):
    for j in range(2, len(TTVdfc.columns)-1):
        bjd = TTVdfc.iloc[i, j]
        bjdTTV = bjd-2457000
        #print("BJD ", TTVdfc.iloc[i,0]," :", bjdTTV)
        TTVCbjd.iat[i, j] = bjdTTV
        TTVCbjd.iat[i,1] = TTVdfc.iloc[i,0]
        TTVCbjd.iat[i,0] = 0
print(TTVCbjd)
TTVCbjd.to_csv('transit_center_c_outer_subtract.csv', index=False)
```
