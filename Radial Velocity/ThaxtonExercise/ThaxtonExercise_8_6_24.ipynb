```python
# Grades Calculator
import random
grades = []
letter_grades = []
students = 26
# Getting Grades of Random Students
for i in range(students):
    grades.append(random.randint(50,100))
print(grades)
# Finding the Average of the "Class"
avg = sum(grades)/len(grades)
print(avg)
# Converting all of the grades from numbers to letters.
for i in range(len(grades)):
    if grades[i]>=90:
        letter_grades.append('A')
    elif 90>grades[i]>=80:
        letter_grades.append('B')
    elif 80>grades[i]>=70:
        letter_grades.append('C')
    elif 70>grades[i]>=60:
        letter_grades.append('D')
    else:
        letter_grades.append('F')
print(letter_grades)
```

```python
# Coin Flipper
import random
heads = 0
tails = 0
flips = 16
# Randomized Coin Flip
for i in range(flips):
    coin_flip = random.randint(1,2)
    if coin_flip == 1:
        heads += 1
    else:
        tails += 1
# Results:
print('Out of ', flips, ' flips, ', heads, ' are heads', ' and ', tails, ' are tails')
```

```python
#OG Sine Fitter
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# Sine function
def sin_function(t, amp, freq, phase, vert):
    return amp * np.sin(freq * t + phase) + vert

# Generate data with noise
t = np.linspace(0, 4 * np.pi, 200)
amp_true, freq_true, phase_true, vert_true = 3.0, 1.5, 0.5, 1.0
y = sin_function(t, amp_true, freq_true, phase_true, vert_true) + np.random.normal(0, 0.4, t.size)

# Fit the sine function to the data
params, params_covariance = curve_fit(sin_function, t, y)
amp_fit, freq_fit, phase_fit, vert_fit = params
print(amp_fit)
print(freq_fit)
print(phase_fit)
print(vert_fit)

# Generate fitted data
y_fit = sin_function(t, amp_fit, freq_fit, phase_fit, vert_fit)

# Error Calculator:
error = y_fit - y
print(y)
print(y_fit)
print(error)

# Plot data and fitted curve
plt.figure(figsize=(12, 6))
plt.plot(t, y, 'o', label='Data', color='red')
plt.plot(t, y_fit, label='Fitted sine curve', color='blue', linewidth=2)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Data Fitting to a Sine Curve')
plt.legend()

# Plot Error
plt.plot(t,error)
```

```python
#Example Phase Folding:
import numpy as np
import matplotlib.pyplot as plt

# Sine function
def sin_function(t, amp, freq, phase, vert):
    return amp * np.sin(freq * t + phase) + vert

# Generate data with noise
times = np.linspace(0, 4 * np.pi, 200)
amp_true, freq_true, phase_true, vert_true = 1.0, 1.0, 1.0, 1.0
rv_measurements = sin_function(times, amp_true, freq_true, phase_true, vert_true) + np.random.normal(0, 0.5, times.size)

# Known period of the signal
period = 6.28

# Calculate phases
phases = (times % period) / period
print(phases)

# Sort by phase
sorted_indices = np.argsort(phases)
sorted_phases = phases[sorted_indices]
sorted_rv_measurements = rv_measurements[sorted_indices]

# Plot the phase-folded data
plt.figure(figsize=(10, 5))
plt.scatter(sorted_phases, sorted_rv_measurements, c='blue', label='Radial Velocity')
plt.xlabel('Phase')
plt.ylabel('Radial Velocity (m/s)')
plt.title('Phase Folded Radial Velocity Plot')
plt.legend()
plt.show()
```

```python
from lmfit import Model
import numpy as np
import matplotlib.pyplot as plt

# Define sine function
def sin_function(t, amp, freq, phase, vert):
    return amp * np.sin(freq * t + phase) + vert

# Generate synthetic data with more noise
t = np.linspace(0, 4 * np.pi, 200)
amp_true, freq_true, phase_true, vert_true = 4.0, 1.5, 0.5, 1.0
y = sin_function(t, amp_true, freq_true, phase_true, vert_true) + np.random.normal(0, 0.5, t.size)

# Create a model based on the sine function
model = Model(sin_function)
params = model.make_params(amp=3.0, freq=1.0, phase=0.5, vert=1.0)

params['amp'].set(min=0, max=10)
params['freq'].set(min=0, max=10)
params['phase'].set(min=-10, max=10)
params['vert'].set(min=0, max=10)


# Perform the fit
result = model.fit(y, params, t=t)
# Print the fit report
print(result.fit_report())

# Plot data and fitted curve
plt.plot(t, y, 'o', label='Data', color='red')
plt.plot(t, result.best_fit, label='Fitted sine curve', color='blue', linewidth=2)
plt.legend()
plt.show()
```

```python
#NEW SINE FITTING, CURVE_FIT WITH A CHI_SQUARE EVALUATOR, updated on Tuesday, at 8:17 PM
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the sine function
def sin_function(t, amp, freq, phase, vert):
    return amp * np.sin(freq * t + phase) + vert

# Generate synthetic data with more noise
t = np.linspace(0, 4 * np.pi, 200)
amp_true, freq_true, phase_true, vert_true = 9.0, 4.5, 3.5, 5.0
y = sin_function(t, amp_true, freq_true, phase_true, vert_true) + np.random.normal(0, 0.5, t.size)
period = float(2*np.pi/1.5)
phase = (t % period) / period


# Initial guesses
initial_guesses = [8.0, 4.0,4.0, 6.0]

# Set the bounds for each parameter: (min, max)
bounds = (
    [0.0, 0.0,0.0, 0.0],  # Lower bounds for amp, freq, phase, vert
    [10.0, 7.0, 3*np.pi, 7.0] # Upper bounds for amp, freq, phase, vert
)

# Perform the initial fit
params, params_covariance = curve_fit(sin_function, phase, y, p0=initial_guesses, bounds=bounds)
amp_fit, freq_fit, phase_fit, vert_fit = params

# Generate fitted data
y_fit = sin_function(phase, amp_fit, freq_fit, phase_fit, vert_fit)
print(amp_fit)
print(freq_fit)
print(phase_fit)
print(vert_fit)


# Error and accuracy calculator
error = y_fit - y
chi_square = np.sum((error)**2 / y)

# Print initial chi_square
print(f"Initial chi_square: {chi_square}")

iteration = 0
max_iterations = 40
previous_chi_square = chi_square

while abs(chi_square) >= 20:
    print(f"Iteration {iteration + 1}: Updating fit...")
    
    # Update initial guesses for the next iteration
    initial_guesses = [amp_fit, freq_fit, phase_fit, vert_fit]
    print(initial_guesses)
    
    # Perform the fit with updated initial guesses
    params, params_covariance = curve_fit(sin_function, t, y, p0=initial_guesses, bounds=bounds)
    amp_fit, freq_fit, phase_fit, vert_fit = params
    print(initial_guesses)
    
    # Generate fitted data and calculate new chi_square
    y_fit = sin_function(phase, amp_fit, freq_fit, phase_fit, vert_fit)
    error = y_fit - y
    chi_square = np.sum((error)**2 / y)
    print(f"Updated chi_square: {chi_square}")
    iteration += 1
    prev_chi_square = chi_square

    if (prev_chi_square - chi_square) < 2:
        print(':)')
        initial_guesses = [param + 4 for param in initial_guesses]
        print(prev_chi_square)
        print(chi_square)
        print(initial_guesses)

print(f"Final chi_square: {chi_square}")

fig, axs = plt.subplots(2)
axs[0].plot(phase, y, 'o', label='Data', color='red')
axs[0].plot(phase, y_fit, label='Fitted sine curve', color='blue', linewidth=2)
axs[0].legend()

axs[1].plot(phase, error, label='Error', color='green')
axs[1].legend()

plt.show()
```
